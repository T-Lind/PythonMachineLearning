```python
from BlackjackEnv import Deck
import matplotlib.pyplot as plt
import numpy as np
from tqdm import tqdm
```

```python
n_iters = 1_000_000
```

```python
two_cards = []
three_cards = []

for _ in tqdm(range(n_iters)):
    deck = Deck()
    value = deck.draw() + deck.draw()
    two_cards.append(value)
    three_cards.append(value + deck.draw())
```

```python
print(f"Two cards mean: {np.mean(two_cards)}, median: {np.median(two_cards)}, st. dev: {np.std(two_cards)}")
print(f"Three cards mean: {np.mean(three_cards)}, median: {np.median(three_cards)}, st. dev: {np.std(three_cards)}")
```

```python
plt.hist(two_cards, 18, color="orange", alpha=0.6, label="Two drawn cards sum")
plt.hist(three_cards, 28, color="red", alpha=0.6, label="Three drawn cards sum")

plt.title("Histogram of blackjack sum probabilities")
plt.xlabel("Hand sum bins")
plt.ylabel("Number of occurances")

plt.legend()
plt.show()
```
